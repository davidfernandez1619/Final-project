```python
def eliminar_espacios_extra(cadena):
    if isinstance(cadena, str):  
        return cadena.strip()
    else:
        return cadena

df = df.applymap(eliminar_espacios_extra)

# We create a function to remove all spaces before and after each value of our dataframe
```
